```python
#line charts and pie charts and bar charts
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
rcParams['figure.figsize']=5,4
sb.set_style('whitegrid')
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
```

```python
#plotting a line chart from pandas object
```

```python
#barcharts
plt.bar(x,y)
```

```python
#to create a horizontal bar chart
plt.barh(x,y)
```

```python
#pie chart
x=[1,2,3,4,5]
plt.pie(x)
plt.show()
```

```python
#saving the chart
plt.savefig('pie_chart.jpeg')
plt.show()
```
