```python
#object oriented plot elements using pandas
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
rcParams['figure.figsize']=5,4
#defining axes,tricks, and grids
x=range(1,10)
y=[1,2,3,4,5,0,1,2,3]
# creating blank figure object
fig=plt.figure()
#addidng the axes
ax=fig.add_axes([.1, .1, 1, 1])
ax.plot(x,y)
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.set_xticks([0,1,2,4,6,8,9])
ax.set_yticks([0,1,2,3,4,5])
ax.plot(x,y)
ax.grid()
ax.plot(x,y)
```

```python
#generate multiple plots in one figure with subplots
fig=plt.figure()
#1 rows 2 columns
fig,(ax1, ax2)=plt.subplots(1,2)
ax1.plot(x)
ax2.plot(x,y)
ax1.plot(x,y)
ax2.plot(x)
```

```python
#plot formatting(defining plot colour,custoizing line styles,setting marker styles)
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
%matplotlib inline
rcParams['figure.figsize']=5,4
sb.set_style('whitegrid')
x=range(1,10)
y=[1,2,3,4,0,4,3,2,1]
#creating bar chart and adjusting colour,bar width and bar allignment
plt.bar(x,y)
#the number of list elements will be number of parameters
wide=[0.5, 0.5, 0.5, 0.9, 0.9, 0.9, 0.5, 0.5, 0.5]
color=['salmon']
plt.bar(x,y,width=wide,color=color,align='center')


```

```python
#the object oriented method
addres='C:/Users/Dhiman/mtcars.csv'
cars=pd.read_csv(addres)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
#isolating mpg variable
mpg=cars.mpg
fig =plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_title('miles per gallon of cars in mtcars')
ax.set_xlabel('car_names')
ax.set_ylabel('miles/gal')


```
